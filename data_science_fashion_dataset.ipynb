```python
# Import pandas
import pandas as pd

# Read the file from Google Drive
file = pd.read_csv('/content/drive/MyDrive/Datasets/Fashion dresses.csv')
file.head(7)
```

```python
file["discount_price"] = file["discount_price"].str.replace("₹","")
file["actual_price"] = file["actual_price"].str.replace("₹","")
```

```python
file['discount_price'] = file["discount_price"].str.replace(',', '').astype(float)
file["actual_price"] = file["actual_price"].str.replace(',', '').astype(float)
```

```python
file['ratings'].unique()
```

```python
file['ratings']=file['ratings'].replace(['FREE'], 0.0)
file['ratings']=file['ratings'].astype(float)
```

```python
file['ratings'].unique()
```

```python
file['no_of_ratings'] = file['no_of_ratings'].astype(str)
file['correct_no_of_ratings'] = pd.Series([file['no_of_ratings'][x][0].isdigit() for x in range(len(file['no_of_ratings']))])

# Drop columns with incorrect 'no_of_ratings'
file = file[file['correct_no_of_ratings'] == True]
file['correct_no_of_ratings'].value_counts()
```

```python
file["no_of_ratings"] = file["no_of_ratings"].str.replace(',', '').astype(float)
```

```python
file.head()
```

```python
file.info()
```

```python
file['ratings'].describe()
```

```python
file=file.dropna(subset=['actual_price','discount_price'])
```

```python
file['manufacturer']=file['name'].str.split(' ').str[0]
cols=file.columns.tolist()
# change the position of manufracturer
cols=['name',
 'manufacturer',
 'main_category',
 'sub_category',
 'image',
 'link',
 'ratings',
 'no_of_ratings',
 'discount_price',
 'actual_price',
 'correct_no_of_ratings']
file=file[cols]
file.head()
```

```python
file[['actual_price','manufacturer']].groupby('manufacturer').mean().round(2).sort_values(by='actual_price',ascending=False)
```

```python
file['discount']=file['actual_price']-file['discount_price']
file['discount_percentage']=round((1-file['discount']/file['actual_price']),2)
file.head()
```

```python
# Import the plotly.express module and assign it to px
import plotly.express as px

# Alternatively, use the full name plotly.express
import plotly.express

# The rest of your code is unchanged
values=file['manufacturer'].value_counts().keys().tolist()[:15]
counts=file['manufacturer'].value_counts().tolist()[:15]
counts,values

plt = px.bar(file, x=values, y=counts, color_discrete_sequence=['GREEN']*len(file))

plt.update_layout(
        plot_bgcolor='WHITE',
        xaxis_title='Name of The Brand',
        yaxis_title='Count',
        title='<b>Popular Electronics Brands</b>'
)

plt.show()
```
